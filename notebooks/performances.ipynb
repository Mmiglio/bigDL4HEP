```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib notebook
```

```python
df = pd.read_csv('/afs/cern.ch/work/m/migliori/public/BDLtest/pythonScripts/hlfTimes.csv')
```

```python
df.head()
```

```python
df = df.sort_values(by='executors', ascending=False)
```

```python
%matplotlib notebook

fig, ax = plt.subplots()
df.loc[df.cores==2].plot('executors', 'elapsedTime', style='-o',
                         label='2 cores per executor', ax=ax)
df.loc[df.cores==6].plot('executors', 'elapsedTime', style='-o',
                         label='6 cores per executor', ax=ax)
plt.xlabel('Number of executors')
plt.ylabel('Training time(s)')
plt.legend(loc="upper right")
plt.grid()
plt.grid()
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.barplot(x='executors', y='throughput', hue='cores', data=df)
plt.ylabel('records/s')
plt.xlabel('Number of executors')
plt.title('Throughput')
plt.legend(loc="upper left")
plt.show()
```
